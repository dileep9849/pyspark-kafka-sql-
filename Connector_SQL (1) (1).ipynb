```python
try:
    conn = msql.connect(host='localhost', database='dileep99', user='root', password='root')
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        cursor.execute('DROP TABLE IF EXISTS health_2;')
        print('Creating table....')

        cursor.execute("CREATE TABLE health_2(id int(200),Gender varchar(255),age int(25),hypertension int(25),heart_disease int(25),ever_married varchar(255),work_type varchar(225),avg_glucose_level int(25),Residence_type varchar(225),bmi int(100),smoking_status varchar(25),stroke int(25))")
        print("Table is created....")
        #loop through the data frame
        for i,row in dileep.iterrows():
            
            sql = "INSERT INTO dileep99.health_2 VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")
            # the connection is not auto committed by default, so we must commit to save our changes
            conn.commit()
except Error as e:
            print("Error while connecting to MySQL", e)
```
